```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

Importing the dataset

```python
dataset=pd.read_excel('stats.xls')
dataset.head()
```

# Preparing the data


The following can be noted in the dataset.
1) Table name, state code, total Urban/Rural is the same throughout the dataset so we will remove them since they donâ€™t have any information value.
2) The town code, district code and Area name hold the same information, so we will be only be keeping Area name and removing the other two.
3) We can also see that the number of unclassified people are very less as compared to the total number of people. We'll remove the last three columns too.


```python
dataset=dataset.drop(dataset.columns[[0,1,3,4,43,44,45]], axis=1)
cols=dataset.columns.tolist()
dataset.head()
```

# State wise analysis

Lets see the total population fo all the states.

```python
state_wise=dataset[(dataset['Age-Group'] == 'All ages')]

fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.barh(state_wise["Area Name"], state_wise["Total - Persons"])
plt.xlabel("Total poulation")
plt.title('Total population of each district of karnataka')
plt.show()
```

We can see from the above graph that population of bangalore is a lot higher than other districts . so we will consider it separately

```python
#Removing banglore for now 
state_wise=dataset[(dataset['Age-Group'] == 'All ages') & (dataset['Area Name'] != 'Bangalore (M Corp.)')]
bangstate=dataset[(dataset['Age-Group'] == 'All ages') & (dataset['Area Name'] == 'Bangalore (M Corp.)') ]

fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.barh(state_wise["Area Name"], state_wise["Total - Persons"])
plt.xlabel("Total poulation")
plt.title('Total population of each district of karnataka')
plt.show()
```

We will normalize the population data and work with percentages.
lets see the total percentage of Literate and Illiterate people for each district.

```python
state_wise.iloc[:, 3:] = 100*(state_wise.iloc[:,3:].div(state_wise.iloc[:,3], axis=0))


#plotting the number of literates and illeterates :illeterates males and females for all states.

# Set position of bar on X axis
barWidth = 0.42
r1 = np.arange(len(state_wise.index))
r2 = [x + barWidth for x in r1]
fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
# Make the plot
plt.bar(r1,state_wise['Illiterate - Persons'], color='Red', width=barWidth, edgecolor='white', label='Illiterate people')
plt.bar(r2,state_wise['Literate - Persons'], color='Green', width=barWidth, edgecolor='white', label='literate people')
plt.xticks([r + barWidth for r in range(len(state_wise.index))], state_wise['Area Name'])
axline = plt.axhline(y = 30, color = 'black', alpha = 0.5,linestyle='--')
plt.xticks(rotation=85)
plt.title('% of literate and illiterate people for each district')
plt.legend()
plt.show()
```

there are 7 districts with literacy rate below 70% so we need to concentrate on these districts more. lets see district wise specs for each of them.

```python
#printing the districts with more that 30% illiteracy rate
state_wise[(state_wise['Illiterate - Persons'] > 30)]['Area Name']
```

since we will be looking at similar observations for each district we will define a common function and use it on all 7 districts.
this function will take the district name as input and return the % of Illiterate males and females from ages of 0-6 to 34-39. 

# District wise Analysis

```python
def statewise_stats(df):
    df.iloc[:, 6:9] = 100*(df.iloc[:,6:9].div(df['Illiterate - Persons'], axis=0))
    barWidth = 0.42
    r1 = np.arange(len(df.index))
    r2 = [x + barWidth for x in r1]
    fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
    # Make the plot
    plt.bar(r1,df['Illiterate - Males'], color='Blue', width=barWidth, edgecolor='white', label='illiterate males')
    plt.bar(r2,df['Illiterate - Females'], color='Cyan', width=barWidth, edgecolor='white', label='illiterate females')
    plt.xticks([r + barWidth for r in range(len(df.index))], df['Age-Group'])
    plt.xticks(rotation=85)
    plt.ylim(0, 100)
    plt.title(df['Area Name'].iloc[0])
    plt.legend()
    plt.show()
```

```python
#lets see the specs for Bijapur. we will only target people till the age of 35 for educational outreach program.
Bijapur=dataset[dataset['Area Name'] == 'Bijapur (CMC)' ].iloc[1:18, :]

statewise_stats(Bijapur)
```

We notice that most illiterate people are of age 18 and above.
lets see the stats for other districts as well.

```python
##lets see the specs for Gulbarga. we will only target people till the age of 35 for educational outreach program.

Gulbarga=dataset[dataset['Area Name'] == 'Gulbarga (M Corp.)' ].iloc[1:18, :]
statewise_stats(Gulbarga)
```

the same seems to be the case for gulbarga as well. Similar to Bijapur the most of illiterate females are of above 18

```python
##lets see the specs for Bidar we will only target people till the age of 35 for educational outreach program.

Bidar=dataset[dataset['Area Name'] == 'Bidar (CMC)' ].iloc[1:18, :]
statewise_stats(Bidar)
```

same as for other districts.

```python
#stats for next two states raichur and Hospet
Raichur=dataset[dataset['Area Name'] == 'Raichur (CMC)' ].iloc[1:18, :]
Hospet=dataset[dataset['Area Name'] == 'Hospet (CMC)' ].iloc[1:18, :]
#plotting the graphs

statewise_stats(Raichur)
statewise_stats(Hospet)

#stats for next two states Bellary and Davanagere
Bellary=dataset[dataset['Area Name'] == 'Bellary (CMC)' ].iloc[1:18, :]
Davanagere=dataset[dataset['Area Name'] == 'Davanagere (CMC)' ].iloc[1:18, :]
#plotting the graphs

statewise_stats(Bellary)
statewise_stats(Davanagere)
```

We can see that most of the districts follow the same trend.

Lets now see what are the education levels like for the 7 distrticts age wise. since our target is to promote primary education we will only consider people of ages 9 10 and 11.

```python
#getting  the list of districts for which illeteracy rate is greater than 30
filter_list=state_wise[(state_wise['Illiterate - Persons'] > 30)]['Area Name'].tolist()
#getting the data for all  those 7 districts
final7=dataset[dataset['Area Name'].isin(filter_list)]
```

now that our dataset is ready we will firt see that stats for people of age 9

```python
age9 = final7[final7['Age-Group'] == 9]
#Normalizing with respect to number of literate people
age9.iloc[:, 9:] = 100*(age9.iloc[:,9:].div(age9.iloc[:,9], axis=0))
display(age9[['Area Name','Educational Level - Below Primary Persons','Educational Level - Primary Persons']])
```

in the above table we can see that at age of 9, only 6% have primary education in Davanagere whereas its 15% in Bidar.
lets see the stats for 10 year olds.

```python
age10 = final7[final7['Age-Group'] == 10]
#Normalizing with respect to number of literate people
age10.iloc[:, 9:] = 100*(age10.iloc[:,9:].div(age10.iloc[:,9], axis=0))
display(age10[['Area Name','Educational Level - Below Primary Persons','Educational Level - Primary Persons']])
```

in the above table we can see that at age of 10, only 29% have primary education in Davanagere whereas its 45% in Bidar.
So we can assume that even though there are considerable amount of literate people most of them are only "below primary" educated for davangere. 
Bidar on the other hand seems to be doing well in terms of "Right Education at Right age"

lets see the stats for 11 year olds.

```python
age11 = final7[final7['Age-Group'] == 11]
#Normalizing with respect to number of literate people
age11.iloc[:, 9:] = 100*(age11.iloc[:,9:].div(age11.iloc[:,9], axis=0))
display(age11[['Area Name','Educational Level - Below Primary Persons','Educational Level - Primary Persons']])
```

In the above, we can aagain see that 77% of the literate 11 year old in bidar have primary education whereas its only 67% for Bellary raichar and Bijapur.
