since we will be looking at similar observations we will define a common fucntion and use it on all 7 districts.
this fuction will take the district name and graph title as the two arguements.

# District wise Analysis

```python
def statewise_stats(df,graph_title):
    df.iloc[:, 6:9] = 100*(df.iloc[:,6:9].div(df['Illiterate - Persons'], axis=0))
    barWidth = 0.42
    r1 = np.arange(len(df.index))
    r2 = [x + barWidth for x in r1]
    fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
    # Make the plot
    plt.bar(r1,df['Illiterate - Males'], color='Blue', width=barWidth, edgecolor='white', label='illiterate males')
    plt.bar(r2,df['Illiterate - Females'], color='Cyan', width=barWidth, edgecolor='white', label='illiterate females')
    plt.xticks([r + barWidth for r in range(len(df.index))], df['Age-Group'])
    plt.xticks(rotation=85)
    plt.ylim(0, 100)
    plt.title(graph_title)
    plt.legend()
    plt.show()
```

```python
#lets see the specs for Bijapur. we will only target people till the age of 40 for educational outreach program.
Bijapur=dataset[dataset['Area Name'] == 'Bijapur (CMC)' ].iloc[1:18, :]

statewise_stats(Bijapur,'Bijapur - Illiterate males Vs Females')
```

We notice that from age of 18 and upwards, the number of illiterate females seems to be increasing.
we can assume that education of girls is only prevalent until 18 since we can see a noticeable difference in the numbers after that age.
lets see the stats for other districts as well.

```python
##lets see the specs for Gulbarga. we will only target people till the age of 40 for educational outreach program.

Gulbarga=dataset[dataset['Area Name'] == 'Gulbarga (M Corp.)' ].iloc[1:18, :]
statewise_stats(Gulbarga,'Gulbarga - Illiterate males Vs Females')
```

the same seems to be the case for gulbarga as well. Similar to Bijapur the % of educated girls is declining after 18.
lets check out other districts.

```python
##lets see the specs for Bidar we will only target people till the age of 40 for educational outreach program.

Bidar=dataset[dataset['Area Name'] == 'Bidar (CMC)' ].iloc[1:18, :]
statewise_stats(Bidar,'Bidar - Illiterate males Vs Females')
```

Again we see that there is a spike in percentages of illeterate women from the age of 18. In this specific District, we can see that the ratio of educated Men Vs Women is almost constant before 18 which might mean that people of that region do not give as much importance to girls education after 18.

```python
#stats for next two states raichur and Hospet
Raichur=dataset[dataset['Area Name'] == 'Raichur (CMC)' ].iloc[1:18, :]
Hospet=dataset[dataset['Area Name'] == 'Hospet (CMC)' ].iloc[1:18, :]
#plotting the graphs

statewise_stats(Raichur,'Raichur - Illiterate males Vs Females')
statewise_stats(Hospet,'Hospet - Illiterate males Vs Females')
```

for the above two districts, we can see that the percentage of illiterate girls are higher right from age 6 which needs to be addressed and remediated.

```python
#stats for next two states Bellary and Davanagere
Bellary=dataset[dataset['Area Name'] == 'Bellary (CMC)' ].iloc[1:18, :]
Davanagere=dataset[dataset['Area Name'] == 'Davanagere (CMC)' ].iloc[1:18, :]
#plotting the graphs

statewise_stats(Bellary,'Bellary - Illiterate males Vs Females')
statewise_stats(Davanagere,'Davanagere - Illiterate males Vs Females')
```

Similar to the previous two districtss, these two districts seem to be following the same trend.
